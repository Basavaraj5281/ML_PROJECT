```python
def check_password(password):
    if  len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False 
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any( char in "!@#$%^&*" for char in password):
        return False
    else:
        return True

print(check_password("Str0ngpwd!"))
print(check_password("Abc@1234"))
```

```python
def Total_costs(cart):
    
    total_cost=0
    for item in cart:
        cost=item["price"]
        quantity=item["quantity"]
        item_cost=cost*quantity
        total_cost=total_cost+item_cost
        

    return total_cost


carts=[
    {"name":"apple","price":0.5,"quantity":10},
    {"name":"ab","price":0.8,"quantity":10},
    {"name":"bc","price":0.9,"quantity":10},
    {"name":"cd","price":0.1,"quantity":10},
    {"name":"de","price":0.4,"quantity":10}
    
    ]

print(Total_costs(carts))
```

```python
'''compute factorial of num using recurssion '''



def factorial(num):

    fact=1
    for i in range(num-1):

        fact=num*fact
        if num<=1:
            break
        else:

            num=num-1
        i=i+1
    return fact


print(factorial(10))
```

```python
class Vectors:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return self.x+other.x,self.y+other.y
    def __sub__(self,other):
        return self.x-other.x,self.y-other.y
    def __mul__(self,num):
        self.num=num
        return self.x*num,self.y*num
v1=Vectors(4,5)
v2=Vectors(5,10)
print(v1+v2)
print(v1*3)
print(v2*4)
print(v1-v2)
```

```python
import math
print("x=real,y is imaginery part")
class Complex_numbers:
    def __init__(self,x,y):
        self.x=x
        self.y=y
        
    def __add__(self,other):
        return complex(self.x+other.x,self.y+other.y)
    def __mul__(self,other):
        real=self.x*other.x-self.y*other.y
        imag=self.x*other.y+self.y*other.x
        return complex(real,imag)

    def __truediv__(self,other):
        real_n=self.x*other.x+self.y*other.y
        imag_n=self.x*other.y-self.y*other.x
        real_d=other.x*other.x+other.y*other.y
        real=real_n/real_d
        imag=imag_n/real_d
        return complex(real,imag)
    def __repr__(self):
        return f"{self.real}+{self.imag}i"
```

```python
z1=Complex_numbers(3,4)
z2=Complex_numbers(5,6)
print(z1+z2)
print(z1*z2)
print(z1/z2)
```
